## Part 2: The Multiple Testing Problem (20 min)

### A thought experiment
Imagine you are a genomics researcher. You have gene expression data for 10,000 genes measured on 40 patients (20 with disease, 20 healthy controls). You run a $t$-test for each gene to see if its expression differs between the two groups.

### Q2.1
If **none** of the 10,000 genes are truly associated with the disease, how many genes would you expect to have $p < 0.05$ by chance alone?

**YOUR ANSWER:**


### Q2.2
Now suppose 100 of the 10,000 genes are truly associated. The remaining 9,900 null genes still produce approximately $9{,}900 \times 0.05 \approx 495$ false positives, plus some additional true positives. If you report all genes with $p < 0.05$ as "significant," roughly what fraction of your reported genes would be false positives?

**YOUR ANSWER:**


### Q2.3
Suggest at least one approach to deal with this problem. You don't need to know the formal name — just describe the idea intuitively.

**YOUR ANSWER:**



````python
# Optional: simulate Q2.1 to confirm your answer
import numpy as np
from scipy import stats

np.random.seed(2026)
n_genes = 10_000
n_per_group = 20
pvals = np.zeros(n_genes)

for i in range(n_genes):
    x = np.random.normal(0, 1, n_per_group)
    y = np.random.normal(0, 1, n_per_group)  # null: same distribution
    _, pvals[i] = stats.ttest_ind(x, y)

n_significant = (pvals < 0.05).sum()
print(f"False positives in {n_genes} pure-null tests: {n_significant}")
print(f"Expected (5% of {n_genes}): {n_genes * 0.05:.0f}")
````
